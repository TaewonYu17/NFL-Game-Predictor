```python
nfl_data['Opp'] = nfl_data['Opp'].replace('ari', 'crd')
nfl_data['Opp'] = nfl_data['Opp'].replace('lar', 'ram')
nfl_data['Opp'] = nfl_data['Opp'].replace('lvr', 'rai')
nfl_data['Opp'] = nfl_data['Opp'].replace('bal', 'rav')
nfl_data['Opp'] = nfl_data['Opp'].replace('hou', 'htx')
nfl_data['Opp'] = nfl_data['Opp'].replace('ten', 'oti')
nfl_data['Opp'] = nfl_data['Opp'].replace('lac', 'sdg')
nfl_data['Opp'] = nfl_data['Opp'].replace('ind', 'clt')
nfl_data['PD'] = (nfl_data['Pts'] - nfl_data['PtsO'])
nfl_data = nfl_data.dropna(subset = ["Week"])
nfl_data["Opp_code"] = nfl_data["Opp"].astype("category").cat.codes
nfl_data["Day_code"] = nfl_data["Day"].astype("category").cat.codes
nfl_data["target"] = (nfl_data["Rslt"] == "W").astype(int)
nfl_data["HorA_code"] = (nfl_data["HorA"] == "@").astype(int)
nfl_data.head(120)
print(nfl_data["Opp"].unique())
print(nfl_data["Team"].unique())
```

```python
BASE_RATING = 1500
K = 20
HOME_FIELD_ADVANTAGE = 50

def calculate_expected_result(rating_a, rating_b, home_field_advantage):
    return 1 / (1 + 10 ** ((rating_b - rating_a + home_field_advantage) / 400))

def calculate_mov_multiplier(point_diff, elo_diff):
    return np.log(abs(point_diff) + 1) * (2.2 / (elo_diff * 0.001 + 2.2))
```

```python
import datetime

current_year = datetime.datetime.now().year
nfl_data['Date'] = pd.to_datetime(nfl_data['Date'])
end_date = f'{current_year}-06-01'
filtered_data = nfl_data[nfl_data['Date'] <= end_date].copy()
filtered_data.sort_values(by=['Date', 'Week'], inplace=True)

history = []
#processed_matchups = set()

for index, row in nfl_data.iterrows():
    team = row['Team']
    opp = row['Opp']
    week = row['Week']
    year = row['Year']
    date = row['Date']
    point_diff = row['PD']
    HorA = row['HorA_code']

#    matchup_key = (year, week, tuple(sorted([team, opp])))
#
#   if matchup_key in processed_matchups:
#      continue  # Skip if already processed

#    processed_matchups.add(matchup_key)
    #teams_on_date = [entry['Team'] for entry in history if entry['Year'] == year and entry['Week'] == week]

    elo_team = elo_ratings.get(team, BASE_RATING)
    elo_opp = elo_ratings.get(opp, BASE_RATING)

    if week == 1:
            
        elo_team += 0
        elo_opp += 0 #placeholder for elo points added for offseason moves
        
    if HorA == 1: home_field_advantage = HOME_FIELD_ADVANTAGE
    else: home_field_advantage = -HOME_FIELD_ADVANTAGE

    if pd.isna(point_diff):
        # For future games, just use the current ELO ratings without updating them
        history.append({
            'Year': row['Year'],
            'Week': row['Week'],
            'Date': row['Date'],
            "Team": team,
            "Opp": opp,
            "Team_Elo_Before": elo_team,
            "Team_Elo_After": elo_team,  # No change for future games
            "Opp_Elo_Before": elo_opp,
            "Opp_Elo_After": elo_opp,    # No change for future games
        })
        print(f"Future Game - Team: {team} Date: {date} Week: {week} Opp: {opp} Team_Elo: {elo_team} Opp_Elo: {elo_opp}")
    else: 
        expected_result = calculate_expected_result(elo_team, elo_opp, HorA)
        mov_multiplier = calculate_mov_multiplier(point_diff, elo_team - elo_opp)

        if point_diff > 0: score = 1
        elif point_diff < 0: score = 0
        else: score = 0.5

        elo_change = K * (score - expected_result) * mov_multiplier

        elo_ratings[team] = elo_team + elo_change
        elo_ratings[opp] = elo_opp - elo_change

        history.append({
            'Year': row['Year'],
            'Week': row['Week'],
            'Date': row['Date'],
            "Team": team,
            "Opp": opp,
            "Team_Elo_Before": elo_team,
            "Team_Elo_After": elo_ratings[team],
            "Opp_Elo_Before": elo_opp,
            "Opp_Elo_After": elo_ratings[opp],
        })
        print(f"Team: {team} Date: {date} Week: {week} Opp: {opp} Team_Elo_Before: {elo_team} Opp_Elo_Before: {elo_opp} Team_Elo_After: {elo_ratings[team]} Opp_Elo_After: {elo_ratings[opp]}")

    """   
        history.append({
            'Year': row['Year'],
            'Week': row['Week'],
            'Date': row['Date'],
            'Week': row['Week'],
            "Team": opp,
            "Opp": team,
            "Team_Elo_Before": elo_opp,
            "Team_Elo_After": elo_ratings[opp],
            "Opp_Elo_Before": elo_team,
            "Opp_Elo_After": elo_ratings[team],
        })
    """
```

```python
nfl_data = nfl_data.merge(
    elo_history_data[
        ["Year", "Week", "Team", "Team_Elo_After", "Opp_Elo_After", "Team_Elo_Before", "Opp_Elo_Before"]
    ],
    on=["Year", "Week", "Team"],
    how="left"
)
nfl_data = nfl_data.sort_values(by=["Team", "Date"])
nfl_data.head(90)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

rf = RandomForestClassifier(n_estimators=100, min_samples_split = 10, random_state = 1)
train = nfl_data[nfl_data['Date'] < '2024-09-01']
test = nfl_data[(nfl_data['Date'] >= ('2024-09-01')) & (nfl_data['Date'] <= ('2025-09-01'))]

predictors = ['HorA_code', 'Opp_code', 'Day_code', 'Team_Elo_Before', 'Opp_Elo_Before', 'Sk', 'TO']
rf.fit(train[predictors], train['target'])
preds = rf.predict(test[predictors])

acc = accuracy_score(test['target'], preds)
acc
```

```python
def rolling_avgs(team, cols, new_cols):
    team = team.sort_values(by="Date")
    rolling_stats = team[cols].rolling(3, closed = "left").mean()
    team[new_cols] = rolling_stats
    team = team.dropna(subset = new_cols)
    team = team[team['Date'] <= pd.Timestamp('2025-09-01')]
    return team
```

```python
cols = ["PD", "Yds", "Rate", "Y/P", "Opp_Elo_Before", "Team_Elo_Before"]
new_cols = [f"{c}_rolling_avg" for c in cols]
new_cols
```

```python
rolling_avgs(team, cols, new_cols)
```

```python
teams_rolling = nfl_data.groupby("Team").apply(lambda x: rolling_avgs(x, cols, new_cols))
teams_rolling = teams_rolling.droplevel("Team")
teams_rolling
```

```python
teams_rolling.index = range(teams_rolling.shape[0])
teams_rolling
```

```python
def make_predictions(data, predictors):
    # Train on all games before 2024-09-01
    train = data[data['Date'] < pd.Timestamp('2024-09-01')]
    
    test = data[
        (data['Date'] >= pd.Timestamp('2024-09-01')) & 
        (data['Date'] <= pd.Timestamp('2025-09-01'))
    ]
    rf.fit(train[predictors], train['target'])
    preds = rf.predict(test[predictors])
    
    combined = pd.DataFrame({
        'actual': test["target"], 
        'prediction': preds,
        'Date': test["Date"],
        'Team': test["Team"],
        'Opp': test["Opp"],
    
    }, index=test.index)
    
    precision = precision_score(test["target"], preds)
    return combined, precision
```
