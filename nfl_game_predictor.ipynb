```python
import pandas as pd
import numpy as np

nfl_data = pd.read_csv('nfl_schedule_stats_2020_2025.csv',index_col=0)
nfl_data.columns
```

```python
nfl_data.sort_values(by=['Team','Year'])
```

```python
print(nfl_data.columns)
```

```python
nfl_data['Opp'] = nfl_data['Opp'].astype('category').cat.codes
nfl_data["HorA"] = nfl_data["HorA"].astype("category").cat.codes
nfl_data['PD'] = (nfl_data['Pts'] - nfl_data['PtsO'])
nfl_data = nfl_data.dropna(subset = ["Week"])
nfl_data.head(20)
```

```python
BASE_RATING = 1500
K = 20
HOME_FIELD_ADVANTAGE = 50

def calculate_expected_result(rating_a, rating_b, home_field_advantage):
    return 1 / (1 + 10 ** ((rating_b - rating_a + home_field_advantage) / 400))

def calculate_mov_multiplier(point_diff, elo_diff):
    return np.log(abs(point_diff) + 1) * (2.2 / (elo_diff * 0.001 + 2.2))
```

```python
elo_ratings = {}
print("PD column info:")
print(nfl_data['PD'].describe())
print("\nFirst 10 PD values:")
print(nfl_data['PD'].head(10))
```

```python
history = []

for index, row in nfl_data.iterrows():
    team = row['Team']
    opp = row['Opp']
    week = row['Week']
    point_diff = row['PD']
    HorA = row['HorA']

    elo_team = elo_ratings.get(team, BASE_RATING)
    elo_opp = elo_ratings.get(opp, BASE_RATING)
    
    if HorA == -1: home_field_advantage = HOME_FIELD_ADVANTAGE
    else: home_field_advantage = -HOME_FIELD_ADVANTAGE

    expected_result = calculate_expected_result(elo_team, elo_opp, HorA)
    mov_multiplier = calculate_mov_multiplier(point_diff, elo_team - elo_opp)

    if point_diff > 0: score = 1
    elif point_diff < 0: score = 0
    else: score = 0.5

    elo_change = K * (score - expected_result) * mov_multiplier

    elo_ratings[team] = elo_team + elo_change
    elo_ratings[opp] = elo_opp - elo_change

    history.append({
        'Year': row['Year'],
        'Week': row['Week'],
        'Date': row['Date'],
        "Team": team,
        "Opp": opp,
        "Team_Elo_Before": elo_team,
        "Opp_Elo_Before": elo_opp,
        "Team_Elo_After": elo_ratings[team],
        "Opp_Elo_After": elo_ratings[opp],
    })
    
    
```

```python
elo_history_data = pd.DataFrame(history)
elo_history_data.head(20)
```

```python
nfl_data = nfl_data.merge(
    elo_history_data[
        ["Year", "Week", "Team", "Opp", "Team_Elo_Before", "Team_Elo_After"]
    ],
    on=["Year", "Week", "Team", "Opp"],
    how="left"
)
nfl_data.head(20)
```
